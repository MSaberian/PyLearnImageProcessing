```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('input/balloons_noisy.png')#, cv2.IMREAD_GRAYSCALE)
result = cv2.medianBlur(image, 5)

cv2.imwrite('output/balloons.jpg', result)
```

```python
image = cv2.imread('input/Medianfilterp.png')#, cv2.IMREAD_GRAYSCALE)
result = cv2.medianBlur(image, 7)

cv2.imwrite('output/Medianfilterp.jpg', result)
```

```python
image = cv2.imread('input/5.png', cv2.IMREAD_GRAYSCALE)
result = cv2.medianBlur(image, 7)
plt.imshow(result, cmap='gray')
plt.show()

cv2.imwrite('output/5.jpg', result)
```
