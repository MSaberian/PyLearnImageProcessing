```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

#sfsv

```python
image = cv2.imread('input\lady_2.jpg', cv2.IMREAD_GRAYSCALE)
hist = cv2.calcHist([image], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

```python
image_low_contrast = cv2.imread('input\old_TV.jpg', cv2.IMREAD_GRAYSCALE)
hist = cv2.calcHist([image_low_contrast], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

```python
image_high_contrast = cv2.equalizeHist(image_low_contrast)

hist = cv2.calcHist([image_low_contrast], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```
