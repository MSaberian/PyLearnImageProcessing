```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('input/Hawkes_Bay_NZ.jpg', cv2.IMREAD_GRAYSCALE)
result = cv2.equalizeHist(image)
hist = cv2.calcHist([image], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
hist = cv2.calcHist([result], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
cv2.imwrite('output/matplotlib.pyplot.jpg', result)
```

```python

plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```
