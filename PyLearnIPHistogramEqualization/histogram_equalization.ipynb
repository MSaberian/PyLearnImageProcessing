```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('input/statue.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
hist = cv2.calcHist([image], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

```python
result = cv2.equalizeHist(image)
plt.imshow(result, cmap='gray')
plt.axis('off')
plt.show()
hist = cv2.calcHist([result], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

```python
claheFilter = cv2.createCLAHE(clipLimit=4, tileGridSize=(40, 40))
result2 = claheFilter.apply(image)
plt.imshow(result2, cmap='gray')
plt.axis('off')
plt.show()
hist = cv2.calcHist([result2], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

```python
resultf = np.hstack((image, result, result2))
plt.figure(figsize=(20,6))
plt.imshow(resultf, cmap='gray')
plt.axis('off')
plt.show()
cv2.imwrite('output/statue.jpg', resultf)
```
