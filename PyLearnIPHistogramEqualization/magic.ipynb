```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('input/1.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
plt.title('Orginal')
plt.show()
```

```python
kernel = np.ones((5,5),np.float32) * 0.04
result_5_4 = cv2.filter2D(image, -1, kernel)
plt.imshow(result_5_4, cmap='gray')
plt.title('Avg-Kernel 5x5 : 0.04')
plt.show()
```

```python
kernel = np.ones((5,5),np.float32) * 5
result_5_5 = cv2.filter2D(image, -1, kernel)
plt.imshow(result_5_5, cmap='gray')
plt.title('Avg-Kernel 5x5 : 5')
plt.show()
```

```python
kernel = np.ones((3,3),np.float32) * 0.04
result_3_4 = cv2.filter2D(image, -1, kernel)
plt.imshow(result_3_4, cmap='gray')
plt.title('Avg-Kernel 3x3 : 0.04')
plt.show()
```
