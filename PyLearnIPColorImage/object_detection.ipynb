```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('input/wolf.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
plt.title('wolf')
plt.axis('off')
plt.show()
```

```python
_, image_thresh = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
plt.imshow(image_thresh, cmap='gray')
plt.title('wolf')
plt.axis('off')
plt.show()
```

```python
contours, _ = cv2.findContours(image_thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

result = image.copy()

for contour in contours:
    if cv2.contourArea(contour) > 1000:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(result, (x,y), (x+w, y+h), (255, 255, 255), 3)

plt.imshow(result, cmap='gray')
plt.title('wolf')
plt.axis('off')
plt.show()
```

```python
image_puma = cv2.imread('input/puma.jpg', cv2.IMREAD_GRAYSCALE)

plt.imshow(image_puma, cmap='gray')
plt.title('wolf')
plt.axis('off')
plt.show()
```

```python
image_puma = 255 - image_puma
plt.imshow(image_puma, cmap='gray')
plt.title('wolf')
plt.axis('off')
plt.show()
```

```python
_, image_puma_thresh = cv2.threshold(image_puma, 127, 255, cv2.THRESH_BINARY)
```

```python
contours, _ = cv2.findContours(image_puma_thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

result = image_puma_thresh.copy()

for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)
    cv2.rectangle(result, (x,y), (x+w, y+h), (127, 127, 127), 2)

plt.imshow(result, cmap='gray')
plt.title('wolf')
plt.axis('off')
plt.show()
```
